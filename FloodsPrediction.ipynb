```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('rainfall.csv')
```

```python
df
```

```python
df = df.drop(['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec','ANNUAL'],axis = 1)
```

```python
TG = df[df['SUBDIVISION']=='TELANGANA']
```

```python
TG.reset_index(inplace = True)
```

```python
TG
```

```python
TG = TG.drop(['index'],axis = 1)
```

```python
TG
```

```python
plt.plot(TG['YEAR'],TG['JAN'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error
import seaborn as sns
```

## JAN MODEL

```python
plt.plot(TG['YEAR'],TG['JAN'])
```

```python
x = TG.YEAR.values.reshape(-1, 1)
y = TG.JAN.values.reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
ModelJan = RandomForestRegressor()
ModelJan.fit(x_train,y_train)
```

```python
y_pred = ModelJan.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
rmse =  np.sqrt(mse)
```

```python
rmse
```

```python

res = ModelJan.predict([[2010],[2025],[2030]])
res
```
